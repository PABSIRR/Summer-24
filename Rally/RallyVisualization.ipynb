```python
def draw_court(ax=None):
    if ax is None:
        ax = plt.gca()
    
    court_color = '#5080B0'
    line_color = 'white'

    # Draw the court background
    ax.set_facecolor('#5080B0')

    # Whole Court (Singles Line X:157 Doubles Line X:177, Baseline Y:455)
    ax.add_patch(patches.Rectangle((-700, -400), 1400, 800))
    ax.add_patch(patches.Rectangle((-455, -210), 910, 420, linewidth=2, edgecolor=line_color, facecolor=court_color))

    # T lines and center lines
    ax.plot([-245, -245], [-157.5, 157.5], color=line_color, linewidth=2)
    ax.plot([245, 245], [-157.5, 157.5], color=line_color, linewidth=2)
    ax.plot([-245, 0], [0, 0], color=line_color, linewidth=2)
    ax.plot([0, 245], [0, 0], color=line_color, linewidth=2)
    
    # Hash lines
    ax.plot([-455, -445], [0, 0], color=line_color, linewidth=2)
    ax.plot([445, 455], [0, 0], color=line_color, linewidth=2)

    # Alley lines
    ax.plot([-455, 450], [-157.5, -157.5], color=line_color, linewidth=2)
    ax.plot([-455, 450], [157.5, 157.5], color=line_color, linewidth=2)

    # Net Line
    ax.plot([0, 0], [-245, 245], color='white', linewidth=2)

    # Set Image Limits
    ax.set_xlim(-700, 700)
    ax.set_ylim(-400, 400)
    
# Create the plot
fig, ax = plt.subplots(figsize=(10.8, 6))
draw_court(ax)

# Remove axes for a cleaner look
ax.axis('off')

# Show the plot
plt.show()
```

## Examples
### Example 1: Shot location for the first point

```python
new_data = data.copy()
new_data = new_data.query('pointNumber == 1')
# new_data = new_data.loc[new_data.pointNumber == 1,:] # Or can use this code

x = new_data['shotLocationX'] * -1 
y = new_data['shotLocationY']

# Create the plot
fig, ax = plt.subplots(figsize=(10.8, 6))
draw_court(ax)

ax.scatter(y, x, c='yellow', linewidth=.02)

# Remove axes for a cleaner look
ax.axis('off')

# Show the plot
plt.show()
```

### Example 2: All of the locations of Elise Wagle's Deuce side serves

```python
new_data = data.copy()
new_data = new_data.query('firstServeIn == 1').query('serverName == "Elise Wagle"').query('side == "Deuce"')

def transform_coordinates(serverSide, x, y):
    if serverSide == 'Near':
        x = x * -1
        y = y * -1
        
    return pd.Series([x, y])

new_data[['firstServeXCoord', 'firstServeYCoord']] = new_data.apply(lambda x: transform_coordinates(x['serverFarNear'], x['firstServeXCoord'], x['firstServeYCoord']), axis = 1)

# x = first_serve_in_data['firstServeXCoord']
x = new_data['firstServeXCoord'] * -1 
y = new_data['firstServeYCoord']

# Create the plot
fig, ax = plt.subplots(figsize=(10.8, 6))
draw_court(ax)

ax.scatter(y, x, c='yellow', linewidth=.02)

# Remove axes for a cleaner look
ax.axis('off')

# Show the plot
plt.show()
```

# Create Animation
### Load packages

```python
import matplotlib.animation as anim
from IPython.display import HTML
```

```python
new_data = data.copy()
new_data = new_data.query('pointNumber == 1')
# new_data = new_data.loc[new_data.pointNumber == 1,:] # Or can use this code

# x = new_data['firstServeXCoord'] * -1 
# y = new_data['firstServeYCoord']

new_data['shotLocationX'].iloc[0] = new_data['firstServeXCoord'].iloc[0]
new_data['shotLocationY'].iloc[0] = new_data['firstServeYCoord'].iloc[0]

x = new_data['shotLocationX'] * -1 
y = new_data['shotLocationY']

# Create the plot
fig, ax = plt.subplots(figsize=(10.8, 6))
draw_court(ax)

scat = ax.scatter(y.iloc[0], x.iloc[0], c='yellow', linewidth=.02)

# Remove axes for a cleaner look
ax.axis('off')

# x = new_data['shotLocationX'] * -1 
# y = new_data['shotLocationY']

# Show the plot
# plt.show()

def update(frame, x, y):
    # for each frame, update the data stored on each artist.
    curr_x = x.iloc[:frame]
    curr_y = y.iloc[:frame]
    
    # update the scatter plot:
    data = np.stack([curr_y, curr_x]).T
    scat.set_offsets(data)
    return [scat]


ani = anim.FuncAnimation(fig=fig, func=update, fargs=(x, y), frames=range(1, 11), interval=500, blit=True)
# plt.show()

# Display the animation in the notebook
HTML(ani.to_jshtml())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

TWOPI = 2*np.pi

fig, ax = plt.subplots()

t = np.arange(0.0, TWOPI, 0.001)
s = np.sin(t)
l = plt.plot(t, s)

ax = plt.axis([0,TWOPI,-1,1])

redDot, = plt.plot([0], [np.sin(0)], 'ro')

def animate(i):
    redDot.set_data(i, np.sin(i))
    return redDot,

# create animation using the animate() function
myAnimation = animation.FuncAnimation(fig, animate, frames=np.arange(0.0, TWOPI, 0.1), \
                                      interval=20, blit=True, repeat=True)

HTML(myAnimation.to_jshtml())
```
