### SwingVision Coord Transformation

```python
# court coordinates
# swing vision - meters, near side center marks (0,0)
# singles court x [-4.1148, 4.1148], y [0, 23.7744]
# doubles court x [-5.485, 5.485]

# our coordinates - center of net (0,0)
# singles court x [-157.5, 157.5], y [-455, 455]

# shot_x = (157.5/4.1148) * swing_x
# shot_y = (455/11.8872) * swing_y + 455
# ratio = 38.2764654418

def transf_x_coord_sv_to_shot(sv_col) :
    return sv_col * 38.2764654418
def transf_y_coord_sv_to_shot(sv_col) :
    return (sv_col - 11.8872) * 38.2764654418
```

### Convert Similar Columns

```python
# shot type booleans
shot_data['shotFhBh'] = shot_data['Stroke'].apply(lambda x: x if x in ['Forehand', 'Backhand'] else '')
shot_data['isSlice'] = shot_data['Spin'].apply(lambda x: '1' if x == 'Slice' else '')
shot_data['isVolley'] = shot_data['Stroke'].apply(lambda x: 1 if x in ['FH Volley', 'BH Volley'] else '')
shot_data['isOverhead'] = shot_data['Stroke'].apply(lambda x: 1 if x == 'Overhead' else '')

# renaming
shot_data.shotInRally = shot_data.Shot
shot_data.side = shot_data['Hit Zone'].str.title()
#shot_data.pointNumber = shot_data.Point

# want to convert swingvision coordinates into our own
shot_data['shotContactX'] = transf_x_coord_sv_to_shot(shot_data['Hit (x)'])
shot_data['shotContactY'] = transf_y_coord_sv_to_shot(shot_data['Hit (y)'])

# Convert ShotContactX and ShotContactY to whole numbers
shot_data['shotContactX'] = np.round(shot_data['shotContactX']).astype(int)
shot_data['shotContactY'] = np.round(shot_data['shotContactY']).astype(int)

shot_data['shotLocationX'] = transf_x_coord_sv_to_shot(shot_data['Bounce (x)'])
shot_data['shotLocationY'] = transf_y_coord_sv_to_shot(shot_data['Bounce (y)'])

# Convert shotLocationX and shotLocationY to whole numbers
shot_data['shotLocationX'] = np.round(shot_data['shotLocationX']).astype(int)
shot_data['shotLocationY'] = np.round(shot_data['shotLocationY']).astype(int)

# shot direction
# down the line --> switches btwn deuce and ad
# crosscourt --> remains on same side

shot_data.shotDirection = np.where(shot_data.shotContactX * shot_data.shotLocationX > 0, "Down the Line", "Crosscourt")
shot_data
```

### Timing Columns 

```python
shot_data['isPointEnd'] = 0
#shot_data.loc[shot_data['Point'].ne(shot_data['Point'].shift(-1)), 'isPointEnd'] = 1
#shot_data.pointEndTime = np.where(shot_data.isPointEnd != '', shot_data['Video Time'], '')

for i in range(len(shot_data.Stroke)):
    # running through every row
    # first shot's a serve, want the player and serving side
    if i == 0:
        cur_server = shot_data.loc[i, "Player"]
        cur_serving_zone = shot_data.loc[i, "Hit Zone"]
    
    if shot_data.loc[i, "Stroke"] == "Serve": # checks if the hit was a serve
        # now want to see if the server or serving zone was changed
        if shot_data.loc[i, "Player"] == cur_server: # same server
            if shot_data.loc[i, "Hit Zone"] != cur_serving_zone: # diff zone
                shot_data.loc[i-1, "isPointEnd"] = 1
                cur_serving_zone = shot_data.loc[i, "Hit Zone"]
        else: # switched servers
            shot_data.loc[i-1, "isPointEnd"] = 1
            cur_serving_zone = shot_data.loc[i, "Hit Zone"]
            cur_server = shot_data.loc[i, "Player"]
shot_data.pointEndTime = np.where(shot_data.isPointEnd != '', shot_data['Video Time'], '')

#shot_data['isPointStart'] = 0
#shot_data.loc[shot_data['Point'].ne(shot_data['Point'].shift()), 'isPointStart'] = 1
shot_data.isPointStart = np.where(shot_data.isPointEnd.shift(1) == 1, 1, 0)
shot_data.loc[0, 'isPointStart'] = 1
shot_data.pointStartTime = np.where(shot_data.Type == 'serve', shot_data['Video Time'], '')

shot_data.loc[0, "pointNumber"] = 1
for i in range(len(shot_data.pointNumber) - 1):
    shot_data.loc[i+1, "pointNumber"] = shot_data.shift(1).loc[i+1,"pointNumber"] + shot_data.shift(1).loc[i+1,"isPointEnd"]

# Point Start and End converted into milleseconds
shot_data['pointStartTime'] = (shot_data['Video Time'] * 60000).astype(int)
shot_data['pointEndTime'] = np.where(shot_data['isPointEnd'] == 1, (shot_data['Video Time'] * 60000).astype(int), '')
```

#### Serving Zones

```python
# serving zones: T, Body, Wide
# Wide: x in [-inf, -105] u [105, inf]
# Body: x in [-105, -52.5] u [52.5, 105]
# T: x in [-52.5, 52.5]

def label_zone(x_coord):
    if x_coord != '':
        x_coord = float(x_coord)
        if x_coord < -105 or x_coord > 105:
            return 'Wide'
        elif -105 <= x_coord <= -52.5 or 52.5 <= x_coord <= 105:
            return 'Body'
        else:
            return 'T'
    return ''

# Convert x_coord to serve zone
shot_data['firstServeZone'] = shot_data['firstServeXCoord'].apply(label_zone)
shot_data['secondServeZone'] = shot_data['secondServeXCoord'].apply(label_zone)
```

### Points

```python
# cols to format: pointScore, gameScore, setScore, pointNumber, isBreakPoint
# iswinner - for eveything reset shot in rally/point end, not double fault, not errors

shot_data['isWinner'] = np.where(
    (shot_data['isPointEnd'] == 1) &
    (shot_data['secondServeIn'] != '0') &
    (shot_data['isErrorWide'] != "1") &
    (shot_data['isErrorNet'] != "1") &
    (shot_data['isErrorLong'] != "1"),
    1,
    ''
)
```

```python
# want to record the score every time a point ends
# points: server - returner
# games: ucla (player1) - opp
# sets: ucla (player1) - opp

# Initialize scores
points = np.zeros(2)  # Points: [server, returner]
games = np.zeros(2)   # Games: [ucla, opp]
sets = np.zeros(2)    # Sets: [ucla, opp]

pt_values = [0, 15, 30, 40]  # Tennis point values
tiebreak_points = np.zeros(2)  # Track points during tiebreak

# Initial scores
shot_data.loc[0, "pointScore"] = f"{pt_values[int(points[0])]} - {pt_values[int(points[1])]}"
shot_data.loc[0, "gameScore"] = f"{int(games[0])} - {int(games[1])}"
shot_data.loc[0, "setScore"] = f"{int(sets[0])} - {int(sets[1])}"

error_cols = [x for x in shot_data.columns if "isError" in x]

for i in range(0, len(shot_data.pointScore) - 1):
    if shot_data.loc[i + 1, "isPointStart"] == 1:
        # Determine point winner
        if shot_data.loc[i, "isWinner"] == 1:
            pt_winner_player_num = '1' if shot_data.loc[i, "Player"] == shot_data.loc[0, "player1Name"] else '2'
        elif any(shot_data.loc[i, error_cols] == "1"):
            pt_winner_player_num = '2' if shot_data.loc[i, "Player"] == shot_data.loc[0, "player1Name"] else '1'
        else:
            print(f"No point recorded at row {i}")
            pt_winner_player_num = None  # Ensure this variable is always defined to prevent missing updates
            
        if pt_winner_player_num is not None:
            didServerWinPt = shot_data.loc[i, "serverName"][-1] == pt_winner_player_num

            if all(games == 6):  # Tiebreak scenario
                if didServerWinPt:
                    tiebreak_points[0] += 1
                else:
                    tiebreak_points[1] += 1

                if any(tiebreak_points >= 7) and abs(tiebreak_points[0] - tiebreak_points[1]) >= 2:
                    sets[np.argmax(tiebreak_points)] += 1
                    games = np.zeros(2)
                    points = np.zeros(2)
                    tiebreak_points = np.zeros(2)
            else:
                if didServerWinPt:
                    points[0] += 1
                else:
                    points[1] += 1

                if any(points > 3) and abs(points[0] - points[1]) >= 1:
                    games[int(pt_winner_player_num) - 1] += 1
                    points = np.zeros(2)

                if any(games > 6) and abs(games[0] - games[1]) >= 2:
                    sets[np.argmax(games)] += 1
                    games = np.zeros(2)

    # Always update scores in DataFrame, even if it's a new point start or no point was recorded
    shot_data.loc[i + 1, "pointScore"] = f"{pt_values[int(points[0])]} - {pt_values[int(points[1])]}"
    shot_data.loc[i + 1, "gameScore"] = f"{int(games[0])} - {int(games[1])}"
    shot_data.loc[i + 1, "setScore"] = f"{int(sets[0])} - {int(sets[1])}"
```
